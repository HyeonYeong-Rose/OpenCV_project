```python
import dlib
import cv2

class PointDist:
    def __init__(self, x, y):
        self.x = x
        self.y = y

def LeftOrRight(landmarks):
    try:
        print("----------------------------")
        dist_left = int(landmarks.part(36).x) - int(landmarks.part(0).x)
        dist_right = int(landmarks.part(16).x) - int(landmarks.part(45).x)
        distance = abs(dist_left / dist_right)
        print("distance : {}".format(distance))
        ratio = [1, 3]
        L = ratio[1] / ratio[0]
        R = ratio[0] / ratio[1]
        if R <= distance <= L:
            print('@_@_@_@_@_@_@_@_@_CENTER_@_@_@_@_@_@_@_@_@')
        elif L < distance:
            print('<<<<<<<<<<<<--------- LEFT')
        elif distance < R:
            print('---------->>>>>>>>> RIGHT')
    except:
        pass

point = []
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

cap = cv2.VideoCapture(0)
flag = True
while flag:
    ret, image_o = cap.read()

    image = cv2.resize(image_o, dsize=(640, 480), interpolation=cv2.INTER_AREA)
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    face_detector = detector(img_gray, 1)

    # 인식된 얼굴 개수
    if(len(face_detector)>1):
        print("부정행위, 사람 2명 이상")
    else:
        face = face_detector[0]
        cv2.rectangle(image, (face.left(), face.top()), (face.right(), face.bottom()),
                      (0, 0, 255), 3)

        landmarks = predictor(image, face)  # 얼굴에서 68개 점 찾기
        #좌,우 판별하기
        LeftOrRight(landmarks)

    cv2.imshow('result', image)

    # wait for keyboard input
    key = cv2.waitKey(33)
    # if esc,
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
